```python
# {:.2f} noktadan sonra 2 basamak alır manasında
text = "{:.2f}, {:.3f}, {:.4f}".format(3.123324, 5.1234352562, 7)
print(text)
```

```python
### bu formatlama yönteminde de string ifadelerin baştan itibaren sınırlınıyoruz (istediklerimizi alıyoruz.) ###
text = "{:.1s} - {:.3s} - {:.4s}".format("3.123324", "5.1234352562", "7.mehmet")
print(text)
```
